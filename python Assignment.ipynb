```python
from datetime import date

class Student:
    def __init__(self, sid, name, m1, m2, m3, dob, fee_paid):
        self.sid = sid
        self.name = name
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3
        self.dob = dob
        self.fee_paid = fee_paid
        self.total_fee = 50000   # assumed total fee

    def calculate_average(self):
        return (self.m1 + self.m2 + self.m3) / 3

    def calculate_cgpa(self):
        avg = self.calculate_average()
        cgpa = avg / 10
        return round(cgpa, 2)

    def calculate_age(self):
        today = date.today()
        age = today.year - self.dob.year - (
            (today.month, today.day) < (self.dob.month, self.dob.day)
        )
        return age

    def fee_balance(self):
        return self.total_fee - self.fee_paid


class College:
    def __init__(self, code, name, location):
        self.code = code
        self.name = name
        self.location = location
        self.students = []

    def register_student(self, student):
        self.students.append(student)

    def display_details(self):
        print("\nCollege Details")
        print("College Code:", self.code)
        print("College Name:", self.name)
        print("Location:", self.location)

        print("\nStudent Academic Report")
        for s in self.students:
            print("\nStudent ID:", s.sid)
            print("Student Name:", s.name)
            print("Marks:", s.m1, s.m2, s.m3)
            print("Average Marks:", s.calculate_average())
            print("CGPA:", s.calculate_cgpa())
            print("Age:", s.calculate_age())
            print("Fee Balance:", s.fee_balance())


# Creating college
college = College("C101", "ABC Engineering College", "Visakhapatnam")

# Registering 3 students
s1 = Student(1, "Rahul", 85, 78, 90, date(2004, 5, 20), 30000)
s2 = Student(2, "Priya", 88, 92, 80, date(2003, 8, 15), 35000)
s3 = Student(3, "Arjun", 75, 82, 79, date(2004, 1, 10), 25000)

college.register_student(s1)
college.register_student(s2)
college.register_student(s3)

# Display report
college.display_details()
```

```python
class PersonalInfo:
    def __init__(self, name, dob, contact, email):
        self.name = name
        self.dob = dob
        self.contact = contact
        self.email = email

    def display_personal(self):
        print("Name:", self.name)
        print("Date of Birth:", self.dob)
        print("Contact Number:", self.contact)
        print("Email:", self.email)


class Education(PersonalInfo):
    def __init__(self, name, dob, contact, email, degree, university, year, cgpa):
        PersonalInfo.__init__(self, name, dob, contact, email)
        self.degree = degree
        self.university = university
        self.year = year
        self.cgpa = cgpa

    def display_education(self):
        print("Degree:", self.degree)
        print("University:", self.university)
        print("Year of Graduation:", self.year)
        print("CGPA:", self.cgpa)


class Experience(Education):
    def __init__(self, name, dob, contact, email, degree, university, year, cgpa,
                 company, role, exp_years, skills):
        Education.__init__(self, name, dob, contact, email, degree, university, year, cgpa)
        self.company = company
        self.role = role
        self.exp_years = exp_years
        self.skills = skills

    def display_experience(self):
        print("Company Name:", self.company)
        print("Job Role:", self.role)
        print("Years of Experience:", self.exp_years)
        print("Skills:", self.skills)


class CandidateProfile(Experience):
    def display_profile(self):
        print("\n----- Candidate Profile -----")
        self.display_personal()
        self.display_education()
        self.display_experience()

    def check_role(self):
        if self.exp_years > 5:
            print("Eligible for Senior Role")
        else:
            print("Eligible for Junior Role")


# Candidate 1
c1 = CandidateProfile(
    "Rahul", "10-05-1998", "9876543210", "rahul@email.com",
    "B.Tech", "JNTU", 2020, 8.2,
    "TCS", "Software Developer", 6, "Python, Java"
)

# Candidate 2
c2 = CandidateProfile(
    "Priya", "15-08-2000", "9123456780", "priya@email.com",
    "B.Sc", "Andhra University", 2022, 7.8,
    "Infosys", "Junior Developer", 3, "Python, SQL"
)

# Display profiles
c1.display_profile()
c1.check_role()

c2.display_profile()
c2.check_role()
```

```python
class MyMath:

    def calculate(self, operation, n):

        if operation == "sum":
            total = n * (n + 1) // 2
            print("Sum of first", n, "natural numbers:", total)

        elif operation == "prime":
            print("First", n, "prime numbers:")
            count = 0
            num = 2
            while count < n:
                is_prime = True
                for i in range(2, num):
                    if num % i == 0:
                        is_prime = False
                        break
                if is_prime:
                    print(num, end=" ")
                    count += 1
                num += 1
            print()

        elif operation == "fibonacci":
            print("Fibonacci series:")
            a, b = 0, 1
            for i in range(n):
                print(a, end=" ")
                a, b = b, a + b
            print()

        elif operation == "factorial":
            fact = 1
            for i in range(1, n + 1):
                fact *= i
            print("Factorial of", n, "=", fact)

        else:
            print("Invalid operation")


# Create object
m = MyMath()

# User input
print("Choose operation: sum / prime / fibonacci / factorial")
op = input("Enter operation: ")
num = int(input("Enter value of n: "))

m.calculate(op, num)
```
